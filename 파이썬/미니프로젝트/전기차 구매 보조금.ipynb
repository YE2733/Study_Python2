```python
import pandas as pd
```

```python
file_path = './한국환경공단_한국환경공단_친환경자동차 정보_전기차구매보조금지급현황정보.csv'
subsidy = pd.read_csv(file_path, encoding='EUC-KR')
```

```python
subsidy.info()
```

```python
subsidy.describe()
```

```python
subsidy.isnull().sum()
```

```python
# 결측치 있는 값 0으로 변환
subsidy = subsidy.fillna(0)
```

```python
subsidy.info()
```

```python
subsidy
```

```python
subsidy[subsidy['시도'] == '경기']
```

```python
# 최대 보조금(만원)/승용으로 내림차순
subsidy2 = subsidy.sort_values('최대 보조금(만원)/승용', ascending=False)
```

```python
# 한국환경공단?
subsidy2[subsidy2['지역구분'] == '한국환경공단']
```

```python
# 최대 보조금(만원)/승용
a = subsidy.groupby('시도')['최대 보조금(만원)/승용'].mean().sort_values(ascending=False).head()
a
```

```python
# 최대 보조금(만원)/초소형
b = subsidy.groupby('시도')['최대 보조금(만원)/초소형'].mean().sort_values(ascending=False).head()
b
```

```python
# 최대 보조금(만원)/화물
subsidy.groupby('시도')['최대 보조금(만원)/화물'].mean().sort_values(ascending=False).head()
```
