```python
import folium
```

```python
#  타입별로 마커 표시하기
default_map = folium.Map(
    # 초기 지도의 중앙점 좌표[위도, 경도]
    location=[37.19949304133392, 126.83083765534775],

    # 지도 스타일
    tiles='openstreetmap',
    # 초기 지도의 스케일
    zoom_start= 11
)

for i in range(len(df2)):
    lat = df2['lat'].iloc[i]
    lng = df2['lng'].iloc[i]
    # print(lat,lng)
    # type = busan_charge_map['충전기타입'].iloc[i]
    # # general : gray, 3
    # # reserve : blue, 5
    # # generalDT : red, 5
    color = 'black'
    # if type == 'AC완속':
    #     fillcolor = 'lightgreen'
    #     size = 5
    # elif type == 'DC콤보':
    #     fillcolor = 'blue'
    #     size = 5
    # else:
    #     fillcolor = 'red'
    #     size = 5
    
    # 지도에 표시할 마킹 모양 등 스타일 지정
    folium.CircleMarker(
        location= [lat, lng],
        # 마커 채우기 색상
        fill_color = 'lightgreen',
        # 마커 채우기 불투명도
        fill_opacity = 1,
        fill=True,
        # 마커 테두리 색상
        color = color,
        # 마커 테두리 굵기
        weight =1,
        # 마커 지름
        radius= 5
    ).add_to(default_map)

default_map
```

```python
default_map.save('화성시_제한구역_지도.html')
```
