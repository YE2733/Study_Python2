```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='Malgun Gothic')
```

```python
file_path = 'C:/Study/Study_Python2/파이썬/8day_포항시버스데이터/files/data_toal.csv'
df_cp = pd.read_csv(file_path)
df_cp
```

```python
df = df_cp.copy()
df
```

```python
# 월, 일, 시 컬럼 추가
df['승차월'] = df['승차시각'].astype(str).str[4:6]
df['승차일'] = df['승차시각'].astype(str).str[6:8]
df['승차시'] = df['승차시각'].astype(str).str[8:10]
```

```python
df['노선명'].value_counts()
```

포항 시민들이 가장 많이 이용하는 버스번호 top5
105번, 107번, 102번, 101번, 200번 입니다.

```python
top_bus = ['105','107','102','101','200']
```

```python
df_top_bus_en = pd.DataFrame()
df_top_bus_ex = pd.DataFrame()
```

```python
for i in top_bus:
    df_en = df[df['노선명'] == i]['승차정류장'].value_counts()[0:1]
    df_top_bus_en = pd.concat([df_top_bus_en, df_en])
    
for i in top_bus:
    df_ex = df[df['노선명'] == i]['하차정류장'].value_counts()[0:1]
    df_top_bus_ex = pd.concat([df_top_bus_ex, df_ex])
```

```python
# 포항 시민들이 가장 많이 이용하는 버스번호 top5의 가장 많이 탑승하는 승차정류장
df_top_bus_en
```

```python
# 포항 시민들이 가장 많이 이용하는 버스번호 top5의 가장 많이 하차하는 하차정류장
df_top_bus_ex
```

시외버스 터미널과 죽도시장을 이용하는 포항 시민 버스 탑승객들이 많습니다.

```python
df_top_bus = pd.DataFrame()
```

```python
for i in top_bus:
    df_temp = df[df['노선명'] == i]
    df_top_bus = pd.concat([df_top_bus, df_temp], ignore_index=True)
    
df_top_bus
```

```python
df_month = df_top_bus.pivot_table(values='승객연령',
                       index='노선명',
                       columns='승차월',
                       aggfunc='count')
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(df_month, annot=True, fmt='.0f', cmap='rocket_r')
plt.title('월별 및 버스노선별 탑승객')
plt.show()
```

```python
df_day = df_top_bus.pivot_table(values='승객연령',
                       index='노선명',
                       columns='승차일',
                       aggfunc='count')
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(df_day, annot=True, fmt='.0f', cmap='rocket_r')
plt.title('일별 및 버스노선별 탑승객')
plt.show()
```

```python
df_hour = df_top_bus.pivot_table(values='승객연령',
                       index='노선명',
                       columns='승차시',
                       aggfunc='count')
df_hour = df_hour.fillna(0)
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(df_hour, annot=True, fmt='.0f', cmap='rocket_r')
plt.title('시간별 및 버스노선별 탑승객')
plt.show()
```
