```python
# 수진자연령, 요양개시일자, 심결본인부담금 
df_month_age = df_20TABLE_org[['수진자연령','요양개시일자','심결본인부담금']]
df_month_age
```

### 연령대 컬럼 만들기

```python
# 연령대 조건 만들기
# 9세이하, 10대~90대, 100대이상
df_9under = df_month_age[df_month_age['수진자연령']<=9]
df_9under = df_9under.reset_index(drop=True)
df_9under['연령대'] = '9세이하'
df_9under
```

```python
# 10대
df_10under = df_month_age[(df_month_age['수진자연령']<=19)&(df_month_age['수진자연령']>=10)]
df_10under = df_10under.reset_index(drop=True)
df_10under['연령대'] = '10대'
df_10under
 
```

```python
# 20대
df_20under = df_month_age[(df_month_age['수진자연령']<=29)&(df_month_age['수진자연령']>=20)]
df_20under = df_20under.reset_index(drop=True)
df_20under['연령대'] = '20대'
df_20under
```

```python
# 30대
df_30under = df_month_age[(df_month_age['수진자연령']<=39)&(df_month_age['수진자연령']>=30)]
df_30under = df_30under.reset_index(drop=True)
df_30under['연령대'] = '30대'
df_30under
```

```python
# 40대
df_40under = df_month_age[(df_month_age['수진자연령']<=49)&(df_month_age['수진자연령']>=40)]
df_40under = df_40under.reset_index(drop=True)
df_40under['연령대'] = '40대'
df_40under
```

```python
# 50대
df_50under = df_month_age[(df_month_age['수진자연령']<=59)&(df_month_age['수진자연령']>=50)]
df_50under = df_50under.reset_index(drop=True)
df_50under['연령대'] = '50대'
df_50under
```

```python
# 60대
df_60under = df_month_age[(df_month_age['수진자연령']<=69)&(df_month_age['수진자연령']>=60)]
df_60under = df_60under.reset_index(drop=True)
df_60under['연령대'] = '60대'
df_60under
```

```python
# 40대
df_70under = df_month_age[(df_month_age['수진자연령']<=79)&(df_month_age['수진자연령']>=70)]
df_70under = df_70under.reset_index(drop=True)
df_70under['연령대'] = '70대'
df_70under
```

```python
# 80대
df_80under = df_month_age[(df_month_age['수진자연령']<=89)&(df_month_age['수진자연령']>=80)]
df_80under = df_80under.reset_index(drop=True)
df_80under['연령대'] = '80대'
df_80under
```

```python
# 90대
df_90under = df_month_age[(df_month_age['수진자연령']<=99)&(df_month_age['수진자연령']>=90)]
df_90under = df_90under.reset_index(drop=True)
df_90under['연령대'] = '90대'
df_90under
```

```python
# 100대 이상 : 데이터 없음
df_month_age[(df_month_age['수진자연령']>=100)]
```
