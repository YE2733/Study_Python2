- 2021년 1월 성별/ 연령별 인구 현황 데이터 가져오기

```python
df = pd.read_csv('C:/Study/Study_Python2/Pandas 실습/people_gender_202101.csv',encoding='cp949', thousands=',')
print(df.head())
```

- 서울 관악구 신림동(1162069500) 정보 확인

```python
df2 = df[df['행정구역'] == '서울특별시 관악구 신림동(1162069500)']
print(df2)
```

- 2021 1월 성별 / 연령별 인구 현황 그래프로 표현

    - 불필요 컬럼 삭제

```python
df2 = df2.drop(['행정구역', '2021년01월_남_총인구수', '2021년01월_남_연령구간인구수','2021년01월_여_총인구수', '2021년01월_여_연령구간인구수'], axis=1)
```

```python
df2.columns
```

    - 남자 데이터 확인

```python
male = df2.iloc[:, 0:11]
male
```

    - 여자 데이터 확인

```python
female = df2.iloc[:, 11:22]
female
```

    - 남자데이터 정리

```python
male2 = male.melt()
male2.drop('variable', axis=1, inplace=True)
male2['value'] = male2['value'].apply(lambda x : x * -1)
male2
```

    - 여자데이터 정리

```python
female2 = female.melt()
female2.drop('variable', axis=1, inplace=True)
female2['value'] = female2['value'].apply(lambda x : x * 1)
female2
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title('서울 신림동 남녀 성별 인구 분포')
plt.barh(range(11), male2['value'].tolist(), label='남자')
plt.barh(range(11), female2['value'].tolist(), label='여자')
plt.yticks(
    range(11),
    ['0-9','10-19','20-29','30-39','40-49','50-59','60-69','70-79','80-89','90-99','100이상'])
plt.xlim(-4500,4500)
plt.xticks(range(-4000,4001,2000),['4000명','2000명','0명','2000명','4000명'])
plt.legend()
plt.show()
```

- 제주도 성별 인구 비율 표현하기

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('C:/Study/Study_Python2/Pandas 실습/people_gender_202101.csv',encoding='cp949', thousands=',')
df
```

```python
df2 = df[df['행정구역'] == '제주특별자치도  (5000000000)']
df2
```

```python
male = df2['2021년01월_남_총인구수']
female = df2['2021년01월_여_총인구수']
plt.rc('font', family='Malgun Gothic')
color = ['magenta', 'cyan']
plt.pie(
    [male.sum(),female.sum()], labels=['남','여'], colors=color, startangle=90, autopct='%.1f%%')

plt.show()
```
